# Import csv file in order to understand the dataset

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Read csv file of gallup polls
csv_file = "csvfile/2019_Great_Jobs_Demonstration_Study_Public_Release_File.csv"
# str = unicode(str, errors='replace')
alldata_df = pd.read_csv(csv_file, header=0, encoding = 'unicode_escape')
alldata_df
```

# Create a dataframe to create the 

```python
# Replace all NaN elements with 0s.
zeroalldata_df = alldata_df.fillna(0)
zeroalldata_df
```

```python
# Create a dataframe with only working respondents
working_data_df = zeroalldata_df.loc[alldata_df['working'] == 1]
working_data_df
```

```python
# Select the columns that will be analyzed
selected_df = working_data_df[['RESPONDENT_ID','Q35_job_category', 'cz_name', 'statename', 'clinton_share', 'trump_share',
                                'total_vote', 'income_level', 'Q50_age', 'Q51_hispanic', 'Q52_white', 'Q52_black',
                                   'Q52_asian', 'Q52_amerind', 'Q52_nativehaw']].copy()
selected_df.head()
```

```python
# Rename the columns

renamed_df = selected_df.rename(columns={"Q35_job_category": "Job Category",
                                        "cz_name": "City",
                                        "statename": "State",
                                        "clinton_share": "Percentage Clinton voters",
                                        "trump_share": "Percentage Trump voters",
                                        "Q50_age": "Age",
                                        "Q51_hispanic": "Hispanic",
                                        "Q52_white": "White/Caucasian",
                                        "Q52_black": "Black or African-American", 
                                         "Q52_asian": "Asian",
                                         "Q52_amerind": "American Indian or Alaska Native",
                                         "Q52_nativehaw": "Native Hawaiian or Pacific Islander",})
renamed_df
```

```python
# Set index
renamed_df.reset_index()
```

# Connect the data frame to SQL

```python
# rds_connection_string = "<insert user name>:<insert password>@localhost:5432/ETL_db"
rds_connection_string = "postgres:postgres@localhost:5432/Project 2"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
renamed_df.to_sql(name='working_respondents', con=engine, if_exists='replace', index=False)
```

```python
pd.read_sql_query('select * from working_respondents', con=engine).head()
```
